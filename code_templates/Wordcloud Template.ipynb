```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import csv
import random
```

```python
def random_color_func(word=None, font_size=None, position=None,  orientation=None, font_path=None, random_state=None):
    h = int(360.0 * 21.0 / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)

    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
with open('LESWordCount - subNY.csv', 'r',newline='\n') as f:
    reader = csv.reader(f)
    next(reader)
    d = {}
    for k,v in reader:
        d[k] = float(v)

 # Generate a word cloud image
# add color_func=random_color_func to call the random color function
wordcloud = WordCloud(background_color='white',max_font_size = 170, width = 750, height = 400).generate_from_frequencies(d) 
```

```python
plt.figure(figsize=(30,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
